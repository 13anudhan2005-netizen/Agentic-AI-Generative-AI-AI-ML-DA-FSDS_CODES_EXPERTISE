```python
x=int(input("a number"))
y=int(input("a number"))
for j in range (1,x+1):
    for k in range(1,y+1):
        if j==1 or j==x:
            print('* ',end=' ')
        else:
            if k==1 or k==y:
                print('* ',end=' ')
            else:
                print(' ',end=' ')
                print()
```
